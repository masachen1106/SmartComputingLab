```python
import radiomics
from radiomics import featureextractor
import SimpleITK as sitk
from PIL import Image 
import pandas as pd
import numpy as np
import os
extractor = featureextractor.RadiomicsFeatureExtractor()

def create_df(img, mask, feature):
    img = sitk.ReadImage(img)
    mask = sitk.ReadImage(mask)
    results = extractor.execute(img, mask, 255)
    key = list(results.keys())
    
    if feature == 'firstorder':
        df = pd.DataFrame(columns=key[22:40] )
    elif feature == 'glcm':
        df = pd.DataFrame(columns=key[40:64] )
    elif feature == 'gldm':
        df = pd.DataFrame(columns=key[64:78] )
    elif feature == 'glrlm':
        df = pd.DataFrame(columns=key[78:94] )
    elif feature == 'glszm':
        df = pd.DataFrame(columns=key[94:110] )       
    elif feature == 'ngtdm':
        df = pd.DataFrame(columns=key[110:] )   
    return df

def get_pyradiomics(img, mask, feature, df):
    img = sitk.ReadImage(img)
    mask = sitk.ReadImage(mask)
    results = extractor.execute(img, mask, 255)
    #key = list(results.keys())
    values = list(results.values())
    
    if feature == 'firstorder':
        df.loc[len(df)] = values[22:40]
    elif feature == 'glcm':
        df.loc[len(df)] = values[40:64]
    elif feature == 'gldm':
        df.loc[len(df)] = values[64:78]
    elif feature == 'glrlm':
        df.loc[len(df)] = values[78:94]
    elif feature == 'glszm':
        df.loc[len(df)] = values[94:110]        
    elif feature == 'ngtdm':
        df.loc[len(df)] = values[110:]
```

```python
import os
from radiomics import featureextractor
import pandas as pd

# 設置原始影像和 mask 影像的資料夾路徑
original_images_folder = r"C:\Users\user\Desktop\FLD\Step1 resize & remove_white_bounder\old_resize_image_20240406"
mask_images_folder = r"C:\Users\user\Desktop\FLD\Step2 Segmentation\Cortext_old_image_20240406"

# 設置 PyRadiomics 的參數
params = {
    'binWidth': 25,  # GLCM 參數
    'resampledPixelSpacing': None,  # 如需重新採樣，可以設置像素間距
    'interpolator': 'sitkBSpline',  # 插值方法
    'verbose': True  # 詳細日誌
}

# 初始化特徵萃取器
extractor = featureextractor.RadiomicsFeatureExtractor(**params)

# 建立空的 DataFrame 用於存儲特徵結果
result_df = pd.DataFrame()

# 遍歷原始影像和 mask 影像資料夾中的文件
for img_file in os.listdir(original_images_folder):
    # 構建原始影像和 mask 影像的完整路徑
    original_image_path = os.path.join(original_images_folder, img_file)
    mask_image_path = os.path.join(mask_images_folder, img_file.replace('.jpg', '_mask.jpg')) 
    print(original_image_path,'\n',mask_image_path)
    
    # 萃取特徵
    result = extractor.execute(original_image_path, mask_image_path)
    
    # 將特徵結果轉換為 DataFrame
    result_series = pd.Series(result)
    img_name = pd.Series({'image_name':img_file})
    result_series = pd.concat([result_series,img_name])
    result_df = result_df.append(result_series, ignore_index=True)

# 將 DataFrame 存儲到 CSV 文件中
result_df.to_csv('feature_extraction_results.csv', index=False)
```
